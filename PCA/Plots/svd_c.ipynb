```r
library(ggplot2)
data=read.csv("/home/naina/Desktop/datamining/svd_c_file.csv")
df=as.data.frame(data)
df <- subset(df, select = -c(X) )
#df
df$X2 <- gsub("b" ,"", df$X2)
df
```

```r
ggplot(df, aes(x=df$X0, y=df$X1, color=df$X2)) + scale_x_continuous("Principal_Component1") +
  scale_y_continuous("Principal_Component2") +
  ggtitle("SVD Plot")+labs(colour="Disease")+
    geom_point(size=2, alpha=0.6)
```
