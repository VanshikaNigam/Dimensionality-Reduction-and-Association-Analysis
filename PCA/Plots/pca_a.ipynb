```r
library(ggplot2)
```

```r
data=read.csv("/home/naina/Desktop/datamining/pca_a_file.csv")
```

```r
df=as.data.frame(data)
```

```r
df
```

```r

df <- subset(df, select = -c(X) )
df$X2 <- gsub("b" ,"", df$X2)
df
```

```r
df
```

```r
ggplot(df, aes(x=df$X1, y=df$X0, color=df$X2)) + scale_x_continuous("Principal_Component1") +
  scale_y_continuous("Principal_Component2") +
  ggtitle("PCA Plot")+
    geom_point(size=2, alpha=0.6)
```

```r
plot(df$X0, df$X1, main="PCA plot", 
  	xlab="Principal_Comp1 ", ylab="Principal_Comp2 ", pch=19,col=df$X2)
```
